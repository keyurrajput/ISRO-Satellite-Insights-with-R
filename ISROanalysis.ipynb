```r
# Load required library for visualization
library(ggplot2)

# Create a line graph to visualize the change in launch mass over time
ggplot(isro_data, aes(x = `Date of Launch`, y = `Launch Mass (kg.)`)) +
  geom_line(color = "royalblue") +
  labs(title = "Change in Launch Mass Over Time", x = "Date of Launch", y = "Launch Mass (kg)") +
  theme_minimal()
```

- **Purpose**: This code generates a line graph.
- **Data**: Launch mass (kg) is plotted over time, with launch dates on the x-axis and mass on the y-axis.
- **Color**: The line color is specified as royal blue.
- **Insight**: It visualizes the trend of launch mass changes over time.

```r
# Count the number of satellites for each purpose
purpose_counts <- table(isro_data$Purpose)

# Create a pie chart
pie(purpose_counts,
    main = "Distribution of Satellite Purpose",
    col = rainbow(length(purpose_counts)),
    cex.main = 1.2,
    cex = 0.8,
    labels = paste(names(purpose_counts), ": ", purpose_counts, sep = ""))

legend("topright", legend = names(purpose_counts), fill = rainbow(length(purpose_counts)))
```

- **Purpose**: This code creates a pie chart.
- **Data**: Counts the number of satellites for each purpose from the "Purpose" column in the dataset.
- **Color**: Uses a rainbow color palette for the pie slices.
- **Insight**: Visualizes the distribution of satellite purposes,

```r
library(ggplot2)

# Count the frequency of each value in Expected Lifetime (yrs.)
lifetime_freq <- table(isro_data$`Expected Lifetime (yrs.)`)

# Convert to data frame
lifetime_df <- as.data.frame(lifetime_freq)

# Rename columns
colnames(lifetime_df) <- c("Lifetime", "Count")

# Plot pie chart
pie_chart <- ggplot(lifetime_df, aes(x = "", y = Count, fill = factor(Lifetime))) +
  geom_bar(stat = "identity") +
  coord_polar("y", start = 0) +
  labs(title = "Expected Lifetime of Satellites",
       fill = "Lifetime (years)",
       x = NULL,
       y = NULL) +
  theme_void() +
  theme(legend.position = "bottom")

# Display pie chart
print(pie_chart)
```

- **Pie chart**: Expected lifetime distribution.
- **Data**: Utilizes "Expected Lifetime (yrs.)".
- **Fill color**: Differentiates lifetime values.
- **Insight**: Displays the distribution of satellite lifetimes in years.

```r
selected_cols <- c("Inclination (degrees)", "Period (minutes)", "Launch Mass (kg.)")

# Check if all columns exist
if (!all(selected_cols %in% colnames(isro_data))) {
  stop("Some of the selected columns are not present in the data!")
}

# Calculate correlation matrix
correlation_matrix <- cor(isro_data[, selected_cols])

# Create heatmap
library(corrplot)  # Load the corrplot package

corrplot(correlation_matrix, method = "number", col = "Red")
```

- **Correlation Heatmap**: Examines relationships.
- **Data**: Uses "Inclination (degrees)", "Period (minutes)", "Launch Mass (kg.)".
- **Insight**: Reveals correlation between selected variables.
